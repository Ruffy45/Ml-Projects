```python
#Checking Array Dimensions
a=np.array(42)
d=np.array([1,2,3,4,5,6])
b = np.array([[1,2,3,6],[1,4,5,6]])
c= np.array([[[1,2,3],[4,5,6]],
             [[1,2,3],[4,5,6]]])
print(a.ndim)
print(d.ndim)
print(b.ndim)
print(c.ndim)


```

```python
arr = np.array([1,2,3,4],ndmin=5)
print(arr)
print('Number of dimensions :', arr.ndim)
```

```python
#operations on of 3rd and 4th element
arr = np.array([1,2,3,4])
print(arr[2]+arr[3])
print(arr[2]-arr[3])
print(arr[2]*arr[3])
print(arr[2]/arr[3])
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
ypoints = np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints = np.array([10,30])
ypoints = np.array([20,300])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xp = np.array([1,8,8,1,1])
yp = np.array([1,1,10,10,1])
plt.plot(xp,yp)
plt.show()
```

```python

yp = np.array([3,6,7,1,3])
plt.plot(yp)
plt.plot(yp,"o")
plt.show()
```

```python
#Dotted red lines o:r
yp = np.array([3,6,7,1,3])
plt.plot(yp)
plt.plot(yp,"o:g")
plt.show()
```

```python
#Dotted green lines o:g with variable dot sizes
yp = np.array([3,6,7,1,3])
plt.plot(yp)
plt.plot(yp,"o:g",ms=10)
plt.title('art')
plt.show()
```

```python
plt.plot(yp,linestyle='dotted')
plt.show()
```

```python
#mfc-marker face color,mec-marker edge color,color
plt.plot(yp,linestyle='dotted',marker='o',mfc='r',mec='b')
plt.show()
```
